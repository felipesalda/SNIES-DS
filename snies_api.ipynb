```python
import requests
import pandas as pd
import numpy as np
import json
```

```python
matriculados = '5wck-szir' # estadística de educación superior 2015-2020 195533.
programas = 'upr9-nkiz' #programas de educación superior del país.
matriculas_dep = '4hrb-y62g' #matriculas por departamento desde 2005.
matriculas_mun = 'y9ga-zwzy' #matriculas por municipio desde 2005.
directivos = 'muyy-6yw9' #directivos de las instituciones de educación superior del país

documentacion = 'https://dev.socrata.com/foundry/www.datos.gov.co/' #documentacion+database
```

```python
def get_data(data):
    # this function gets the data from the API through the request module
    response = requests.get(
    'https://www.datos.gov.co/resource/{data}.json?$limit={limit}'.format(data=data,limit=200000)
    ).text

    response = json.loads(response)
    return response
```

## matriculados

```python
df= pd.DataFrame(data=get_data(matriculados))
df
```

```python
df.columns
```

```python
df = df.rename(columns={'c_digo_de_la_instituci_n':'código_IES', 'ies_padre':'ies_padre',
       'instituci_n_de_educaci_n_superior_ies':'IES', 'principal_oseccional':'principal_oseccional',
       'id_sector':'sector', 'id_caracter':'caracter', 'c_digo_del_departamento_ies':'cod_departamento_ies',
       'departamento_de_domicilio_de_la_ies':'dep_dom_ies', 'c_digo_del_municipio_ies':'cod_municipio',
       'municipio_dedomicilio_de_la_ies':'municipio_IES', 'c_digo_snies_delprograma':'SNIES_programa',
       'programa_acad_mico':'programa', 'id_nivel':'nivel', 'id_nivel_formacion':'nivel_formacion',
       'id_metodologia':'metodologia', 'id_area':'area', 'id_nucleo':'id_nucleo',
       'n_cleo_b_sico_del_conocimiento_nbc':'NBC',
       'c_digo_del_departamento_programa':'cod_departamento_programa',
       'departamento_de_oferta_del_programa':'departamento_programa', 'c_digo_del_municipio_programa':'cod_municipio_programa',
       'municipio_de_oferta_del_programa':'municipio_programa', 'id_g_nero':'genero', 'a_o':'año', 'semestre':'semestre',
       'matriculados_2015':'matriculados'})
```

```python
df
```

```python
#replace values

df['nivel'] = df['nivel'].map({'1':'Pregrado',
                             '2':'Posgrado',
                             np.nan:'Sin clasificar'},na_action=None)

df['area'] = df['area'].map({'1':'Agronomía, veterinaria y afines',
                             '2':'Bellas artes',
                             '3':'Ciencias de la educación',
                             '4':'Ciencias de la salud',
                             '5':'Ciencias sociales y humanas',
                             '6':'Economía, administración, contaduría y afines',
                             '8':'Ingeniería, arquitectura, urbanismo y afines',
                             '9':'Matemáticas y ciencias naturales',
                             np.nan:'Sin clasificar'},na_action=None)

df['nivel_formacion'] = df['nivel_formacion'].map({'1':'Especialización universitaria',
                             '2':'Maestría',
                             '3':'Doctorado',
                             '4':'Formación técnica profesional',
                             '5':'Tecnológica',
                             '6':'Universitaria',
                             '7':'Especialización técnico profesional',
                             '8':'Especialización tecnológica',
                             '10':'Especialización médico quirúrgica',
                             np.nan:'Sin clasificar'},na_action=None)

df['metodologia'] = df['metodologia'].map({'1':'Presencial',
                             '2':'Distancia (Tradicional)',
                             '3':'Distancia (Virtual)',
                             '4':'Presencial-Virtual',
                             np.nan:'Sin clasificar'},na_action=None)

df['genero'] = df['genero'].map({'1':'Masculino',
                             '2':'Femenino',
                             np.nan:'Sin clasificar'},na_action=None)

df['sector'] = df['sector'].map({'1':'Oficial',
                             '2':'Privada',
                             np.nan:'Sin clasificar'},na_action=None)

df['caracter'] = df['caracter'].map({'1':'Institución Técnica Profesional',
                             '2':'Institución Tecnológica',
                             '3':'Institución Universitaria/Escuela Tecnológica',
                             '4':'Universidad',
                             np.nan:'Sin clasificar'},na_action=None)
```

```python
# data filtering example
df[(df.area =='Bellas artes')]
```

```python
df.to_csv('data_processed/matriculados.csv', encoding='utf-8')
```
